```python
import pandas as pd
from sqlalchemy import create_engine
from config import POSTGRES

# Build connection string
conn_str = (
    f"postgresql://{POSTGRES['user']}:{POSTGRES['password']}"
    f"@{POSTGRES['host']}:{POSTGRES['port']}/{POSTGRES['database']}"
)
engine = create_engine(conn_str)
print("Connected to database!")
```

```python
neg_men = pd.read_sql("""
    SELECT * FROM matches_men
    WHERE home_score < 0 OR away_score < 0
""", engine)
neg_men
```

```python
neg_women = pd.read_sql("""
    SELECT * FROM matches_women
    WHERE home_score < 0 OR away_score < 0
""", engine)
neg_women
```

```python
high_scores_men = pd.read_sql("""
    SELECT * FROM matches_men
    WHERE home_score > 20 OR away_score > 20
""", engine)
high_scores_men
```

```python
high_scores_women = pd.read_sql("""
    SELECT * FROM matches_women
    WHERE home_score > 20 OR away_score > 20
""", engine)
high_scores_women
```

```python
neutral_issue_men = pd.read_sql("""
    SELECT * FROM matches_men
    WHERE neutral = TRUE AND home_team = country
""", engine)
neutral_issue_men
```

```python
neutral_issue_women = pd.read_sql("""
    SELECT * FROM matches_women
    WHERE neutral = TRUE AND home_team = country
""", engine)
neutral_issue_women
```
